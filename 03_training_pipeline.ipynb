### Defining the model

Gradients tell us how much and in which direction a model’s output (or loss) changes when we slightly change a parameter (like weights).  

   """requires_grad=True  
    tells PyTorch to track gradients  
    Needed for backpropagation  
    Allows .backward() to compute gradients for this tensor"""

```python
class SimpleNeuralNetworct():

  def __init__(self, X): #constructor , X is a training set

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True) #initialize weights randomly

    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X): # (w*X+b) *sigmoid
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 30
```

### Training Pipeline

```python
# create model
model = SimpleNeuralNetworct(X_train_tensor) #input data

# define loop
for epoch in range(epochs):
   #Four steps for traing the model
  # 1 forward pass (z=w*X+b)
  y_pred = model.forward(X_train_tensor)

  # 2 loss calculate
  loss = model.loss_function(y_pred, y_train_tensor)

  # 3 backward pass
  loss.backward()

  # 4 parameters update
  with torch.no_grad(): #Gradient tracking off
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # zero gradients
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  # print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
# model evaluation
with torch.no_grad(): #At the evalution time no need of backpropagation so we off the graded tracking
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.5).float() # if y_pres >0.5 then 1 other 0
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')

  print(y_pred)
  print(y_test_tensor)
```
